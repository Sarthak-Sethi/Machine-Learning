```python
#importing the matplot,pyplot for plotting graphs
import matplotlib.pyplot as plt
# x has the x co-ordinates of all the points and y has y- co ordinates
x =[1,3,10,4]
y =[2,6,1,8]
# scatter function plots all points on graph
plt.scatter(x,y)
# show id sued to display the graph
plt.show()
```

```python
# TO JOIN THE POINTS FORMED USE PLT.PLOT
plt.plot(x,y)
plt.show()
```

```python
# usinG SCATTER AND PLOT TOGETHER
plt.scatter(x,y)
plt.plot(x,y)
plt.show()
```

```python
# PLOTTING CUBE ROOT
import numpy as np
# np.arange will make an array starting from 0 till 20 and step of 0.1
arr = np.arange(0,20,0.1)
# since it is an np array arr**3 will cube each element in array
cubearr = arr**3
# plotting the graph for the same
plt.plot(arr,cubearr)
plt.show()
```

```python
# customising graphs
# labelling x axis , y axis and title
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Cube root graph")
# makes grid lines at back 
plt.grid()
# axis takes 4 parameters as list in it 
# starting index of x-axis =0
# ending index of x-axis =20
# starting index of y-axis =0
# ending index of y-axis =8000
plt.axis([0,20,0,8000])
plt.plot(arr,cubearr)
plt.show()
```

```python
# LABEELING THE LINES 
squarearr = arr**2
plt.plot(arr,squarearr,label="X^2")
plt.plot(arr,cubearr,label="X^3")
plt.legend()
plt.show()
```

```python
# ADDING TEXT TO YOUR GRAPH
plt.text(3,500,"point of sepratation")
plt.plot(arr,squarearr,label="X^2")
plt.plot(arr,cubearr,label="X^3")
plt.show()
```

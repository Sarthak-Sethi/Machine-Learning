```python
import matplotlib.pyplot as plt
sizes=[3,5,8,7,6,7]
lables=["A","B","C","D","E","F"]
colors=["blue","red","green","pink","yellow","orange"]
plt.title("pie chart")
plt.pie(sizes,colors=colors,labels=lables,autopct="%.2f%%")
plt.axis("equal")
plt.show()
```

```python
# to change the direction of flow of pie chart
# as above it is anticlockwise 
# to make it clockwise we added counterclocwiese = False rest is same
plt.pie(sizes,colors=colors,labels=lables,autopct="%.2f%%",counterclock=False)
plt.axis("equal")
plt.show()
```

```python
# sometimes we want a particular pie to be showed different from other for that we can use explode
explode=[0.2,0,0.5,0,0,0]
plt.pie(sizes,colors=colors,labels=lables,autopct="%.2f%%",explode=explode)
plt.axis("equal")
plt.show()
```

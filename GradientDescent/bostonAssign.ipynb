### cost function
formula of cost = (y-m1x1+m2x2+......)**2
x and y are used as fetched 

```python
def cost(data, m):
    total_cost = 0
    N = len(data)
    for i in range(N):
        x = x_in_arr[i]
        y = y_in_arr[i]
        total_cost += (1/N)*((y - (m*x).sum())**2)
    return total_cost
```

### step gradient 
first loop is for every row second id for columns , we are saying ki har row ke liye m_slope nikalo <br>
now for every slope for each row multiply it by alpha using loop 3

```python
def step_gradient(data,alpha,m):
    m_slope = np.zeros(14)
    N = len(data)
    for i in range(N):
        x = x_in_arr[i]
        y = y_in_arr[i]
        for j in range(14):
            m_slope[j]+=(-2/N)*(y - (m*x).sum())*x[j]
    
    # loop 3
    for j in range(14):
        m[j] = m[j]- alpha*m_slope[j]
    return m
```

```python
def GD(data, num_iterations,alpha):
    m = np.zeros(14)
    for i in range(num_iterations):
        m = step_gradient(data,alpha,m)
        print(i, " Cost: ", cost(data, m))
    return m    
```

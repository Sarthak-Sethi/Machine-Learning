### importing required packages

```python
from sklearn import datasets
from sklearn.linear_model import LogisticRegression
import pandas as pd
import numpy as np
from pandas_profiling import ProfileReport
import seaborn as sns
```

### loading the x and y

```python
data = pd.read_csv("file:///D:/ML/LogisticRegression/training_titanic_x_y_train.csv")
data.head()
```

```python
data.drop(["Name","Ticket","Cabin"],inplace=True,axis=1)
data.head()
```

```python
data.drop("Fare",inplace=True,axis=1)
data.head()
```

### converting gender to int so that we can perform operation as we cannot delete it , 
### as we have seen that it affects
Converting sex with male = 0 and female = 1

```python
def getNumber(str):
    if str=="male":
        return 0
    else:
        return 1
data["gender"]=data["Sex"].apply(getNumber)
del data["Sex"]
data.head()
```

### handling the nan in the age by replacing it with mean on the basis of class

```python
Pclassone = data[data['Pclass']==1]
p1 = Pclassone["Age"].mean()
print(p1)
Pclasstwo = data[data['Pclass']==2]
p2 = Pclasstwo["Age"].mean()
print(p2)
Pclassthree = data[data['Pclass']==3]
p3 = Pclassthree["Age"].mean()
print(p3)
def handle_nan(data):
    age = data[0]
    pclass = data[1]
    if pd.isnull(age):
        if pclass==1:
            return p1
        elif pclass==2:
            return p2
        else:
            return p3
    else:
        return age
    
data["age"] = data[["Age","Pclass"]].apply(handle_nan,axis=1)
del data["Age"]
data["age_age"] = data["age"]**2
data.head()
```

### droping all the nan as only 1 is left that too in embarked and then checking that all if good 

```python
data.dropna(inplace=True)
```

```python
data.isnull().sum()
```

### since embarked is still object and has 3 distinct values 
1. S 
2. C 
3. Q <br>
replacing s with 0 c with 1 and q with 2

```python
def getNumber(str):
    if str=="S":
        return 0
    elif str=="C":
        return 1
    else :
        return 2
data["emarked"]=data["Embarked"].apply(getNumber)
del data["Embarked"]
data.head()
```

```python
data["survived"] = data["Survived"]
del data["Survived"]
data.head()
```

### fetching x and y from data

```python
train_data = data.values
train_data
x = train_data[:,0:7]
y = train_data[:,7]
```

### LogisticRegression

```python
Algo = LogisticRegression(max_iter=5000)
Algo.fit(x,y)
```

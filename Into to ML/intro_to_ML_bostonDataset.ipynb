```python
from sklearn import datasets
# sklearn contains lots of ml algo and sample datasets one of them is boston
boston = datasets.load_boston()
boston
```

```python
# x and y are numpy array and x is the data and y is the output of that data
x = boston.data
y = boston.target
type(x)
```

```python
import pandas as pd
# we want to convert our data into a dataframe so below command
df = pd.DataFrame(x)
# if you print boston.feature_names as we see in data above printed , 
# it is nothing but the column names so seetting it as column name for the same
df.columns = boston.feature_names
df
```

```python
# now to see which  column means what we need its description for which we will seee boston.DESC 
# as we can see in data printed above
df.describe()
# by describe we can analyse that in CHA s 75% of the values are 0 , so these things can be seen using describe
```

```python
from sklearn import model_selection as ms
# model_selection is used to split the input data randomly 
# why randomly ?
# if we have 100 inputs data we want to train our model using 80 and then test for its accuracy using remaining 20
# so if we leave last 20 or first 20 thats not good ( see copy for more)
# so to split data randomly we use train_test_split function of model_selection
ms.train_test_split(x,y)
```

```python
# the train_test_split function will split the data randomly into train and test data as we see above it gives
# 4 arrays as output which are
#x_train  = input for training model
#y_train  = output for training model
#x_test  = input for testing model
#y_test  = output for testing model

x_train,x_test,y_train,y_test = ms.train_test_split(x,y)
# by seeing the shape of it
print(x_train.shape)
print(x_test.shape)
```

```python
from sklearn.linear_model import LinearRegression 
# as sklearn has different algo one of them is linear regression
algo1 = LinearRegression()
# the above line is creating the object of algo1 which is LinearRegression
algo1.fit(x_train,y_train)
# fit function takes x input and y output and is used to train the model 
y_predicted_by_our_algo = algo1.predict(x_test)
# y_predicted_by_our_algo has the output predicted by our model / algo when we pass x_test which is the remaining 25 % we kept
# for testing how accurate our algo is
y_predicted_by_our_algo
# this below array has they predicted by our algo 
# now we want to see how accurate our output is for which we will draw graph between y predicted by our algo and y test
```

```python
import matplotlib.pyplot as plt
# plotting graph between predcited and y test now if y_test is 10 y predicted should be 10 that means the line
# y = x should be plotted 
# which is not very accurate as seen below
# but things are not too far away so our algo is working fine
plt.scatter(y_test,y_predicted_by_our_algo)
plt.show()
```

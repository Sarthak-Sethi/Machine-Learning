```python
from sklearn import datasets
import pandas as pd
# load diabtets
diab = datasets.load_diabetes()
# fetch input and output in x and y as x me input and y me output
x = diab.data
y = diab.target
# convert input into table
df = pd.DataFrame(x)
# changing column names
df.columns = diab.feature_names
df.head()
```

```python
# viewinf description to understand the input columns
diab.DESCR
```

```python
# checking for null and anaylysing/ cleaning data
# most of data is clean in sklern samples so kuch krna nhi padha
df.describe()
```

```python
from sklearn import model_selection as ms
# this will break the input and output in 75 and 25 ratio with 75 for training and 25 in testing 
# returns 4 parameters as seen
xtrain,xtest,ytrain,ytest  = ms.train_test_split(x,y)
print(xtrain.shape)
print(ytrain.shape)
print(xtest.shape)
print(ytest.shape)
# as we can see below data is divided into 331 rows for training and y train is just the output for the same which is 1d array
# so it is (331,)
# simiplary for testing it is 111 rows
```

```python
from sklearn.linear_model import LinearRegression
# linear regression is algo used 
import matplotlib.pyplot as plt
# created algo object
algo = LinearRegression()
# passng training data to let model train himself
algo.fit(xtrain,ytrain)
# getting and storing predicted data by our model for our testing data which is ytest in ypred
ypred = algo.predict(xtest)
# plotting graph for both ytest and ypred it should be equal for accurate reuslts
plt.scatter(ytest,ypred)
plt.show()
```

```python
#RESETTING INDEX WITHOUT ADDING THE INDEX COLUMN
iris = pd.read_csv("file:///D:/ML/grades.csv")
iris.drop([0,5,10],inplace=True)
iris.reset_index(drop=True,inplace=True)
iris
# by writing drop=True it will not add index column as above cell
# in the below output the origirnal 0 5 and 10 labels are missing and it is then reset_index
```

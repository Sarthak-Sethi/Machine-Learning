```python
import pandas as pd
import numpy as np
iris = pd.read_csv("file:///D:/ML/grades.csv")
# the below command will replace the data from 2nd row till 4th row and from 2nd col to 3rd col by a constant of numpy 
# which is nan
iris.iloc[2:5,2:4]=np.nan
iris.head(7)
```

```python
# removing nan from table
table_in_which_all_na_are_dropped =iris.dropna()
table_in_which_all_na_are_dropped.head(6)
```

```python
iris.columns = ['lname','fname','location','t1','t2','t3','t4','final','grade']
iris = iris.astype({'t1':float ,'final': float})
iris.dtypes
```
